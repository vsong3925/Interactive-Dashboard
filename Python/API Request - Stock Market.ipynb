```python
import pandas as pd
import requests
from config import api_key
import os
```

```python
ticker_file = "../data/ticker_list.csv"
```

```python
ticker_list = pd.read_csv(ticker_file)
ticker_list.head(10)
```

```python
tickers = ticker_list["Quandl_Code"][0:1000]

company_name = []
company_id = []
company_code = []
oldest_available_date = []
newest_available_date = []
close_data = []

for t in tickers:
    
    ticker = t
    column_index = "3"
    start_date = "2021-02-01"
    end_date = "2021-02-02"
    
    url_time = f"https://www.quandl.com/api/v3/datasets/{ticker}?start_date={start_date}&end_date={end_date}&api_key={api_key}"
   
    stock = requests.get(url_time).json()

    try:
        company_name.append(stock["dataset"]["name"])
        company_id.append(stock["dataset"]["id"])
        company_code.append(stock["dataset"]["dataset_code"])
        oldest_available_date.append(stock["dataset"]["oldest_available_date"])
        newest_available_date.append(stock["dataset"]["newest_available_date"])
        close_data.append(stock["dataset"]["data"][0][5])

        company_name_df = pd.DataFrame(company_name)
        company_id_df = pd.DataFrame(company_id)
        company_code_df = pd.DataFrame(company_code)
        oldest_available_date_df = pd.DataFrame(oldest_available_date)
        newest_available_date_df = pd.DataFrame(newest_available_date)
        close_data_df = pd.DataFrame(close_data)

        print(f'----------{t}--------')
    except:
        print(f'------{t}------- not found')
        pass
```

```python
stock_df = pd.concat([company_name_df, company_id_df, company_code_df, 
                      oldest_available_date_df, newest_available_date_df, 
                      close_data_df], axis=1)
stock_df.head(30)
```
