```python
import pandas as pd
import requests
from config import api_key
import os
```

## Read csv

```python
ticker_file = "../static/data/ticker_list_original.csv"
```

```python
ticker_list = pd.read_csv(ticker_file)
ticker_list.head()
```

```python
stocks_industries = "../static/data/industries.xlsx"
industries = pd.read_excel(stocks_industries)
industries.head()
```

## Merge

```python
merged_df = industries.merge(ticker_list, on="Ticker")
merged_df1 = merged_df.rename(columns = {"Name_x": "Name"})
merged_df1
```

```python
ticker_df = merged_df1[["Ticker", "Quandl_Code", "Name", "Industry", "Exchange"]]
ticker_df.to_csv("../static/data/tickerlist_industries_condensed.csv")
```
